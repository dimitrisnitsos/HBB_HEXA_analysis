```python
#Import required libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

<h5>Analysis for HBB<h5>

```python
df = pd.read_excel("HBB_HEXA_anzalone_data.xlsx", sheet_name='HBB')
df
```

```python
df = pd.read_excel("HBB_HEXA_anzalone_data.xlsx", sheet_name='HBB')
# Calculate average for % correct edit (w/o indels), rep 1 - 3
df2 = df
df['Mean of % correct edit (w/o indels) '] = df.iloc[:, [1,2,3]].mean(axis=1)
df = df.drop(['% correct edit (w/o indels), rep 1','% correct edit (w/o indels), rep 2','% correct edit (w/o indels), rep 3'],axis=1)
 # Calculate the average for % indels, rep 1 - 3
df['Mean of % indels'] = df2.iloc[:, [4,5,6]].mean(axis=1)
df = df.drop(['% indels, rep 1','% indels, rep 2','% indels, rep 3'],axis=1)
```

```python
df_HBB = df.drop('Mean of % indels', axis = 1)
df_HBB
```

```python
# Export to excel HBB
# forming dataframe
data = pd.DataFrame(df_HBB) 
  
# storing into the excel file
data.to_excel("HBB.xlsx")
```
